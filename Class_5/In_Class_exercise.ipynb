```python
# Now, we could drop the location column since the location does not provide too much information for us,
# the reason why the location is meaningless is that 968 out of 1000 people are from different regions.
# Include the location column would be wasteful and resulted in high memory consumption
# drop the location column and then save dt as a new dataset called dt2
dt2 = dt.drop(columns=['location'])
dt2.head()
```

```python
dt2.info()
```

```python
# Drop the signup_date and last_purcahse_date column sice we already have the signup_year, signup_month
# And last_purchase_year, and last_purchase_month
dt2.drop(columns=['signup_date', 'last_purchase_date'], inplace=True)
```

```python
# Now, our dataset dt_new contains only numeric value, which is ready to conduct PCA analysis.
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
dt2_std = scaler.fit_transform(dt2)
```

```python
# Perform PCA
pca = PCA()
dt2_pca = pca.fit_transform(dt2_std)
```

```python
# The transformed data is an array, convert it back into a dataframe
dt2_pca = pd.DataFrame(dt2_pca, columns=[f'PC{i+1}' for i in range(len(dt2.columns))])
```

```python
# Print the explained variance ratio
print('Explained variance ratio:', pca.explained_variance_ratio_)

# Print the cumulative explained variance ratio
cumsum_variance = np.cumsum(pca.explained_variance_ratio_)
print('Cumulative explained variance ratio:', cumsum_variance)
```

```python
# Show the first few rows of transformed dataframe
dt2_pca.head()
```

```python
# Let's do the same, but now let's reduce to 9 components
# Perform PCA
pca = PCA(n_components = 9) #or n_components can also specify how many information you want the PCA analysis to include (has to between 0 and 1)
dt2_pca = pca.fit_transform(dt2_std)

# The transformed data is an array, convert it back into a dataframe
dt2_pca = pd.DataFrame(dt2_pca)

# Print the explained variance ratio
print('Explained variance ratio:', pca.explained_variance_ratio_)

# Print the cumulative explained variance ratio
cumsum_variance = np.cumsum(pca.explained_variance_ratio_)
print('Cumulative explained variance ratio:', cumsum_variance)

# Show the first few rows of transformed dataframe
dt2_pca.head()
```
